```python
# ── CELL 4: Multi-Document Loader ──────────────────────────
# Handles: PDFs, Emails (.eml), Chat Logs (.txt / .csv)

def load_pdfs(folder):
    docs = []
    for path in glob.glob(os.path.join(folder, "**/*.pdf"), recursive=True):
        try:
            loader = PyPDFLoader(path)
            pages  = loader.load()
            for page in pages:
                page.metadata["source"] = os.path.basename(path)
                page.metadata["type"]   = "PDF"
            docs.extend(pages)
            print(f"  ✅ PDF:      {os.path.basename(path)} ({len(pages)} pages)")
        except Exception as e:
            print(f"  ❌ PDF fail: {os.path.basename(path)} — {e}")
    return docs


def load_emails(folder):
    docs = []
    for path in glob.glob(os.path.join(folder, "**/*.eml"), recursive=True):
        try:
            with open(path, "rb") as f:
                msg = email.message_from_bytes(f.read())
            body = ""
            if msg.is_multipart():
                for part in msg.walk():
                    if part.get_content_type() == "text/plain":
                        body += part.get_payload(decode=True).decode(errors="replace")
            else:
                body = msg.get_payload(decode=True).decode(errors="replace")
            subject = msg.get("Subject", "No Subject")
            sender  = msg.get("From",    "Unknown")
            date    = msg.get("Date",    "Unknown")
            content = f"Email Subject: {subject}\nFrom: {sender}\nDate: {date}\n\n{body.strip()}"
            docs.append(Document(
                page_content=content,
                metadata={"source": os.path.basename(path), "type": "Email",
                          "subject": subject, "sender": sender}
            ))
            print(f"  ✅ Email:    {os.path.basename(path)}")
        except Exception as e:
            print(f"  ❌ Email fail: {os.path.basename(path)} — {e}")
    return docs


def load_chat_logs(folder):
    docs = []
    for path in glob.glob(os.path.join(folder, "**/*.txt"), recursive=True):
        try:
            with open(path, "r", encoding="utf-8", errors="replace") as f:
                raw = f.read().strip()
            docs.append(Document(
                page_content=f"Chat Log: {os.path.basename(path)}\n\n{raw}",
                metadata={"source": os.path.basename(path), "type": "Chat Log"}
            ))
            print(f"  ✅ Chat TXT: {os.path.basename(path)}")
        except Exception as e:
            print(f"  ❌ TXT fail: {os.path.basename(path)} — {e}")

    for path in glob.glob(os.path.join(folder, "**/*.csv"), recursive=True):
        try:
            with open(path, "r", encoding="utf-8", errors="replace") as f:
                reader = csv.DictReader(f)
                rows   = list(reader)
            if not rows:
                continue
            headers     = {h.lower(): h for h in rows[0].keys()}
            speaker_col = next((headers[k] for k in headers if any(x in k for x in ["speaker","sender","from","user","agent"])), None)
            message_col = next((headers[k] for k in headers if any(x in k for x in ["message","text","content","body"])), None)
            time_col    = next((headers[k] for k in headers if any(x in k for x in ["time","date","ts"])), None)
            lines = []
            for row in rows:
                ts  = f"[{row[time_col]}] "   if time_col    else ""
                spk = f"{row[speaker_col]}: " if speaker_col else ""
                msg = row[message_col]          if message_col else " | ".join(row.values())
                lines.append(f"{ts}{spk}{msg}")
            docs.append(Document(
                page_content=f"Chat Log: {os.path.basename(path)}\n\n" + "\n".join(lines),
                metadata={"source": os.path.basename(path), "type": "Chat Log (CSV)"}
            ))
            print(f"  ✅ Chat CSV: {os.path.basename(path)}")
        except Exception as e:
            print(f"  ❌ CSV fail: {os.path.basename(path)} — {e}")
    return docs


print("📂 Loading documents...\n")
all_docs = []
all_docs.extend(load_pdfs(DOCS_FOLDER))
all_docs.extend(load_emails(DOCS_FOLDER))
all_docs.extend(load_chat_logs(DOCS_FOLDER))
print(f"\n📊 Total documents loaded: {len(all_docs)}")


def smart_split(docs):
    chunks = []
    for doc in docs:
        doc_type = doc.metadata.get("type", "PDF")
        splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000 if doc_type == "PDF" else 500,
            chunk_overlap=150 if doc_type == "PDF" else 50
        )
        chunks.extend(splitter.split_documents([doc]))
    return chunks

chunks = smart_split(all_docs)
print(f"✂️  Total chunks: {len(chunks)}")
```

```python
# ── CELL 5: Embeddings & Vector Store ──────────────────────
print("🔍 Building vector store (this takes ~30s)...")
embeddings   = OpenAIEmbeddings()
vector_store = FAISS.from_documents(chunks, embeddings)
retriever    = vector_store.as_retriever(search_kwargs={"k": 4})
print("✅ Vector store ready!")
```

```python
# ── CELL 6: QA Chain ───────────────────────────────────────
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)

def ask_question(query):
    result  = qa_chain({"query": query})
    answer  = result["result"]
    sources = [doc.metadata.get("source", "Unknown") for doc in result["source_documents"]]
    return answer, sources

def extract_text_from_response(response):
    if isinstance(response, str):
        return response
    if isinstance(response, dict):
        return response.get("result", str(response))
    return str(response)

print("✅ QA chain ready! You can now launch the chatbot.")
```
